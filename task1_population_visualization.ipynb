```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(
    "dataset/API_SP.POP.TOTL_DS2_en_csv_v2_34.csv",
    skiprows=4
)

# Select 2022 population
pop_2022 = df[['Country Name', '2022']].dropna()

# Top 10 countries
top_10 = pop_2022.sort_values(by='2022', ascending=False).head(10)

# Set professional style
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12,6))

bars = sns.barplot(
    x='2022',
    y='Country Name',
    hue='Country Name',
    data=top_10,
    palette='viridis',
    legend=False
)

plt.title(
    "Top 10 Most Populous Countries in 2022",
    fontsize=15,
    fontweight='bold'
)
plt.xlabel("Population", fontsize=12)
plt.ylabel("Country", fontsize=12)

# Add population labels
for i, value in enumerate(top_10['2022']):
    plt.text(value, i, f"{int(value/1e6)}M", va='center', fontsize=10)

plt.tight_layout()
plt.savefig("images/top10_population_2022.png", dpi=300)
plt.show()
```
